```python
#Open and Read the system generated rdf file
with open('systemnvt.rdf.0.5','r') as f:
    system_data05=f.readlines()
with open('systemnvt.rdf.1','r') as f:
    system_data1=f.readlines()
with open('systemnvt.rdf.1.5','r') as f:
    system_data15=f.readlines()
with open('systemnvt.rdf.2','r') as f:
    system_data2=f.readlines()
with open('systemnvt.rdf.4','r') as f:
    system_data4=f.readlines()
with open('systemnvt.rdf.6','r') as f:
    system_data6=f.readlines()
with open('systemnvt.rdf.8','r') as f:
    system_data8=f.readlines()
with open('systemnvt.rdf.10','r') as f:
    system_data10=f.readlines()
with open('systemnvt.rdf.12','r') as f:
    system_data12=f.readlines()
with open('systemnvt.rdf.14','r') as f:
    system_data14=f.readlines()
```

```python
#getting the system Coordinates at different epsilon for 1-2 interaction
def get_system_coordinates(system_data,start,end):
    x=[]
    y=[]
    for i in system_data[start:end]:
        m=i.split()
        x.append(float(m[1]))
        y.append(float(m[4]))
    return x,y
x05,y05=get_system_coordinates(system_data05,4,104)
x1,y1=get_system_coordinates(system_data1,4,104)
x15,y15=get_system_coordinates(system_data15,4,104)
x2,y2=get_system_coordinates(system_data2,4,104)
x4,y4=get_system_coordinates(system_data4,4,104)
x6,y6=get_system_coordinates(system_data6,4,104)
x8,y8=get_system_coordinates(system_data8,4,104)
x10,y10=get_system_coordinates(system_data10,4,104)
x12,y12=get_system_coordinates(system_data12,4,104)
x14,y14=get_system_coordinates(system_data14,4,104)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15, 8))
plt.plot(x05,y05 ,'b--', label="Eps=0.5")
plt.plot(x1,y1,'r--',label="Eps=1.0")
plt.plot(x15,y15,label="Eps=1.5")
plt.plot(x2,y2 ,label="Eps=2.0")
plt.plot(x4,y4,label="Eps=4.0")
plt.plot(x6,y6,label="Eps=6.0")
plt.plot(x8,y8,label="Eps=8.0")
plt.plot(x10,y10,'b',label="Eps=10")
plt.plot(x12,y12,'g',label="Eps=12")
plt.plot(x14,y14,'r',label="Eps=14")

plt.legend()
plt.show()
print("System Generated Rdf Plot at different Epsilon for 1-2 Interaction")
```
